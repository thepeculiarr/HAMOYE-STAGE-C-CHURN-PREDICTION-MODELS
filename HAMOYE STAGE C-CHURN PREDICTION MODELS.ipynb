```python
#Importing Relavant Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading 

df = pd.read_csv('C:\\Users\\PECU\\Documents\\WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
#Copying raw data

dff = df.copy()
dff
```

```python
#DATA CLEANING AND MAPPING

dff['SeniorCitizen'] = dff['SeniorCitizen'].astype('category')
```

```python
#converting Total Charges to Numeric

dff['TotalCharges'] = pd.to_numeric(dff['TotalCharges'], errors = 'coerce') 
```

```python
pd.to_numeric(dff['TotalCharges'], errors = 'coerce')
```

```python
dff.isnull().sum()
```

```python
dff['TotalCharges'].fillna(0, inplace = True)
```

```python
#Getting info of data after Claeaning

dff.info()
```

```python
#Mapping 'Churn'

dff['Churn'] = dff['Churn'].map({'Yes':1 , 'No':0})
dff
```

```python
#Dropping customer  iD

dff.drop('customerID', axis =1, inplace=True)
```

```python
#getting columns

dff.columns
```

```python
#train-test-split

from sklearn.model_selection import train_test_split

X = dff.drop(['Churn'],axis=1)
y = dff['Churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#List of categorical and numerical features

categorical = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                    'PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
                   'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling',
                   'PaymentMethod']

categorical
```

```python
numerical= ['tenure', 'TotalCharges', 'MonthlyCharges']
numerical
```

```python
#Encoding train ategorical features and turning into a a dataframe after encoding

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)

X_train_encoded = ohe.fit_transform(X_train[categorical])
X_train_encoded
```

```python
#getting feature names as headers for train

categorical_name = list(ohe.get_feature_names_out(input_features=categorical))
```

```python
#Converting to Dataframe

train_encoded = pd.DataFrame(X_train_encoded, columns = categorical_name)
train_encoded
```

```python
#Standard Scaling of train numerical features and turning into a dataframe after encoding

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train[numerical])
X_train_scaled = scaler.transform(X_train[numerical])
X_train_scaled
```

```python
#Converting to Dataframe

train_scaled = pd.DataFrame(X_train_scaled, columns= numerical)
train_scaled
```

```python
#Concatenating train dataframe

train_dataframe = pd.concat([train_encoded, train_scaled], axis = 1)
train_dataframe
```

```python
#Encoding test categorical features and turning into a a dataframe after encoding

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output = False)

X_test_encoded = ohe.fit_transform(X_test[categorical])
X_test_encoded
```

```python
#getting feature names as headers for test

categorical_name = list(ohe.get_feature_names_out(input_features=categorical))
```

```python
#Converting to Dataframe

test_encoded = pd.DataFrame(X_test_encoded, columns = categorical_name)
test_encoded
```

```python
#Standard Scaling of test numerical features and turning iinto a dataframe after encoding

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_test[numerical])
X_test_scaled = scaler.transform(X_test[numerical])
X_test_scaled
```

```python
#Converting to Dataframe

test_scaled = pd.DataFrame(X_test_scaled, columns= numerical)
test_scaled
```

```python
#Concatenatingg Test dataframe

test_dataframe = pd.concat([test_encoded, test_scaled], axis = 1)
test_dataframe
```

```python
#Model training[EXTRAtreesClassifier, RandomForestCllassifier,XGBoost, LGBMClassifier] and Accuracy scores

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import accuracy_score

#Extra Trees Classifier and accuracy score
ext = ExtraTreesClassifier(random_state = 1)
ext.fit(train_dataframe,y_train)
ext_pred = ext.predict(test_dataframe)
print(accuracy_score(ext_pred,y_test))
```

```python
#RandomForestClassifier and accuracy score

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state = 1)
rf.fit(train_dataframe, y_train)
rf_pred = rf.predict(test_dataframe)
print(accuracy_score(rf_pred,y_test))
```

```python
#XGBoost Classifier and accuracy score

from xgboost import XGBClassifier
xgb = XGBClassifier(random_state = 1)
xgb.fit(train_dataframe, y_train)
xgb_pred = xgb.predict(test_dataframe)
print(accuracy_score(xgb_pred,y_test))
```

```python
#Lightgbm Classifier and accuracy score

from lightgbm import LGBMClassifier
lgb = LGBMClassifier(random_state = 1)
lgb.fit(train_dataframe, y_train)
lgb_pred= lgb.predict(test_dataframe)
print(accuracy_score(lgb_pred,y_test))
```
